```python
# Tamanhos de matrizes
tamanho = [200, 300, 900, 1300]
```

```python
# Tempos coletados ao rodar os diferentes programas
tempo_py = [0.66, ((2.35+2.32+2.42+2.45+2.38)/5), 86.24, 297.84]

tempo_cpp = [0.0793687, 0.260289, 7.82261, 22.4972]

tempo_parel = [0.076083, 0.252622, 7.16009, 25.2324]
```

```python
import matplotlib.pyplot as plt
plt.plot(tamanho, tempo_py, label='Python')
plt.plot(tamanho, tempo_cpp, label='C++')
plt.plot(tamanho, tempo_parel, label='Paralelo')
plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo de execução (s)')
plt.title('Tempo de execução por tamanho de matriz na minha máquina')
plt.legend()
plt.show()
```

```python
# Tamanhos de matrizes
tamanho_cluster = [150, 300, 450, 600, 750, 900, 1050, 1200, 1350, 1500]
```

```python
# Tempos coletados ao rodar os diferentes programas
tempo_py_cluster = [0.28, 2.42, 8.24, 19.75, 38.63, 81.55, 114.41, 167.37, 254.50, 358.97]

tempo_cpp_cluster = [0.0179637, 0.145289, 0.498536, 1.18088, 2.32644, 4.01814, 6.40574, 9.87287, 13.9671, 19.1717]

tempo_parel_cluster = [0.00639997, 0.0448452 , 0.148571, 0.350432, 0.665703, 1.1671, 1.85466, 2.79932, 3.94277, 5.43097]
```

```python
plt.plot(tamanho_cluster, tempo_py_cluster, label='Python')
plt.plot(tamanho_cluster, tempo_cpp_cluster, label='C++')
plt.plot(tamanho_cluster, tempo_parel_cluster, label='Paralelo')
plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo de execução (s)')
plt.title('Tempo de execução por tamanho de matriz no Cluster')
plt.legend()
plt.show()
```

## Discuta o impacto do ambiente HPC no desempenho das operações e como o paralelismo pode escalar em um cluster de múltiplos nós.

O ambiente HPC aumenta significativamente a eficiência de aplicações/códigos computacionalmente intensivos. Isso é evidente em operações como cálculos matriciais, onde o aumento do tamanho do problema não se torna imediatamente o principal fator limitante da eficiência na resolução. No entanto, mesmo em um ambiente HPC, não é possível contornar as diferenças de eficiência entre diferentes linguagens de programação.
